```python
import matplotlib.pyplot as plt

# Veriler
labels = ["SciFi", "Drama", "Gerilim", "Komedi"]
sizes = [30, 25, 20, 25]  # Yüzdelik dağılım (toplam 100 olmak zorunda değil)
explode = (0.1, 0, 0, 0)  # SciFi biraz dışarı çıksın

# Grafik oluşturma
plt.figure()
plt.pie(sizes, 
        explode=explode, 
        labels=labels, 
        autopct='%1.1f%%', 
        shadow=True, 
        startangle=90)

plt.title("Film Türü Dağılımı")
plt.show()
```

```python
import matplotlib.pyplot as plt

categories = ["Mat", "Fizik", "Kimya", "Biyoloji"]
values = [85,90,78,88]

plt.bar(categories, values)

plt.title("Öğrenci puanları")
plt.xlabel("Dersler")
plt.ylabel("Puanlar")

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

veri = np.random.normal(70, 10, 100)

plt.hist(veri, bins= 10)

plt.title("notların dağılımı")
plt.xlabel("not")
plt.ylabel("frekans")

plt.show()
```

```python
import matplotlib.pyplot as plt

enerji_kaynakları = [35, 25, 18, 12, 10]

etiketler = [
    "lightskyblue",
    "CORNFLOWERBLUE",
    "CORNSILK",
    "CADETBLUE",
    "CYAN"
]

renkler = [
    "lightskyblue",
    "CORNFLOWERBLUE",
    "CORNSILK",
    "CADETBLUE",
    "CYAN"
]

plt.pie(
    enerji_kaynakları,
    labels=etiketler, 
    autopct="", #İçindeki yazı
    colors=renkler 
)

plt.title("Shades of Blue") # başlık

plt.show()
```
