```python
# Kişisel Bilgiler Formu

# Kullanıcıdan değer alyoruz.
isim = input("İsminizi Girin: ")
soyisim = input("Soyisminizi Girin: ")
yas = int(input("Yaş: ")) # yaş intager olduğu için int ile aldık.
sehir = input("Şehir: ")
meslek = input("Meslek: ")

# 5 yıl sonunda olacak yaşı ve toplam harf sayısını buluyoruz.
bes_yıl_sonunda = yas + 5
toplam_harf = len(isim) + len(soyisim) # len fonksiyonu kaç harf/karakter olduğunu belirtir.

# Girilen değerleri ekrana yazdırıyoruz.
print(f"Kişisel Bilgiler")
print(f"İsim Soyisim  : {isim} {soyisim}")
print(f"Yaşınız  : {yas}")
print(f"Şehir  : {sehir}")
print(f"Meslek  : {meslek}")
print(f"5 yıl sonraki yaşınız  : {bes_yıl_sonunda}")
print(f"İsim Soyisim Toplam Harf Harf Sayısı   : {toplam_harf}")
```
