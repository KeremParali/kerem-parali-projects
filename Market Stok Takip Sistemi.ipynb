```python
# Market stok takip sistemi ödevi;
""" 
    Her ürün için dictionary olsun (ad,fiyat,stok)
    Tüm ürünler listede tutulsun    
    En az 3 ürün(for döngüsü ile listele)
    Stok 5ten az ise kritik stok uyarısı (if koşulu)
    Toplam stok değeri hesaplması (fiyat x stok)
"""

# ---- KEREM PARALI --- NO = 300125024 ----


# ürünlerin listesi
market = []  # Tüm ürünler bir list içinde tutuluyor.

# ürünleri ekliyorum 
urun1 = {"ad": "Ananas", "fiyat": 150, "stok": 10}  # her ürün için dictioanary oluşturduk
urun2 = {"ad": "Şeker", "fiyat": 227, "stok": 4}
urun3 = {"ad": "Süt", "fiyat": 84, "stok": 2}

market.append(urun1)  # append() listeye yeni bir eleman eklemek için kullanılır
market.append(urun2)  # burada oluşturduğumuz ürünleri market listesine ekliyoruz
market.append(urun3)

print("MARKET STOK LİSTESİ")
print("---------------------")

toplam_stok_degeri = 0  # Toplam stok değerini hesaplamak için başlangıç değeri

# ürünleri listeleme
for i in market:  # 'in' ifadesi listenin içindeki elemanları tek tek denemesini sağlar
                  # 'i' her döngüde listedeki bir ürünü temsil eder
    
    ad = i["ad"]      # i = dictioanary demektir
    fiyat = i["fiyat"]  # dictioanary içinden fiyat değerini alıyorum
    stok = i["stok"]    # dictioanary den stok değerini alıyorum. x < 5 ise kritik diyecek


    
    print("Ürün:", ad) # alt alta yazdırıyorum
    print("Fiyat:", fiyat)
    print("Stok:", stok)
    
    # kritik stok uyarısı
    if stok < 5:  # stok sayısı 5ten küçüse uyarı vermek için if koşulu kullandım (if = eğer)
        print("Kritik stok seviyesinde !!!") # gösterilen mesaj
    
    if stok > 5:  
        print("Yeterli stok seviyesi") 

    print("---------------------")
    
    # toplam değer hesaplaması
    toplam_stok_degeri = toplam_stok_degeri + (fiyat * stok)  

print("Toplam stok değeri:", toplam_stok_degeri) # toplam değeri yazdırıyorum
```
