```python
"""
ŞİFRE KONTROL SİSTEMİ ÖDEVİ

En az 8 karakter olmalı
En az 1 büyük harf içermeli(isupper)
En az bir küçük harf(islower)
En az 1 rakam(isdigit)
Her fonksiyona docstring eklenmeli


# ÖDEV3
# KEREM PARALI --- 300125024

"""

"""
--- DÖNGÜLER ---
"""
def uzunluk_kontrol(sifre):   # for döngüsü
    
    # Şifrenin en az 8 karakter olup olmadığını kontrol eder.
    
    return len(sifre) >= 8          # len karakter uzunluğunu hesaplamak için kullandık

def kontrol_buyuk(sifre):   # for döngüsü
    
    # Şifre içinde en az bir büyük harf olup olmadığını isupper() ile kontrol ettik.
    
    for karakter in sifre:      # döngü başlattık 
        if karakter.isupper():      # büyük harf aratıyoruz
            return True     # bulursa true bulamazsa false döndürecek 
    return False 

def kontrol_kucuk(sifre):    # küçük harfi kontrol ediyoruz
    
    # Şifre içinde en az bir küçük harf olup olmadığnı islower() ile bakıyoruz
    
    for karakter in sifre:      # döngü başlattık
        if karakter.islower():      # en az 1 küçük harf arıyoruz
            return True        # bulursa true bulamazsa false 
    return False

def kontrol_rakam(sifre):   # rakam kontrolü
    
    # Şifre içinde en az bir rakam olup olmadığını isdigit() ile kontrol eder.
    
    for karakter in sifre:  
        if karakter.isdigit(): # en az 1 rakam kontrolü için isdigit kullandık
            return True
    return False

"""
--- ŞİFRE KONTROL ---
"""
def sifre_kontrol(sifre):
    
    # List oluşturdum ve hepsini içine alıp kontrollerini sağladım
    
    eksik_kurallar = [] # list

    if not uzunluk_kontrol(sifre): 
        eksik_kurallar.append("Şifreniz en az 8 karakter olmalıdır. ❌") # verilen hata mesajları

    if not kontrol_buyuk(sifre):
        eksik_kurallar.append("Şifreniz en az 1 büyük harf içermelidir. ❌") # append listin sonuna eleman eklemek için kullanılır

    if not kontrol_kucuk(sifre):
        eksik_kurallar.append("Şifreniz en az 1 küçük harf içermelidir. ❌")

    if not kontrol_rakam(sifre):
        eksik_kurallar.append("Şifre en az 1 rakam içermelidir. ❌")
                                                        
    return eksik_kurallar # şartları sağlayana kadar bu mesajları gönderecek


"""
--- ONAY / HATA ---
"""

# kullanıcıdan şifre alma kısmı
sifre = input("Şifre giriniz: ") 

sonuc = sifre_kontrol(sifre) # şartları sağlıyormu kontrol ediyoruz

if len(sonuc) == 0: # sağlıyorsa onay yazıları atıcak (== eşitlik kontrolü)
    print("Giriş Yapıldı")
    print("8 Karakter ✅")
    print("1 Büyük Harf ✅")
    print("1 Küçük Harf ✅")
    print("1 Rakam ✅")

else: # eğer
    print("Giriş Başarısız Şartlar Sağlanmadı") # başarısızsa gösterilen mesaj
    print("Şartlar:")

    if uzunluk_kontrol(sifre): # burada şartı sağladıysa onay sağlamadıysa error verdiriyorum
        print("8 Karakter ✅")
    else:
        print("8 Karakter ❌")

    if kontrol_buyuk(sifre):  # büyük harf şartı varsa onay yoksa error verdirdim
        print("1 Büyük Harf ✅")
    else:
        print("1 Büyük Harf ❌")

    if kontrol_kucuk(sifre): # küçük harf kontrol varsa onay yoksa error verdirdim
        print("1 Küçük Harf ✅")
    else:
        print("1 Küçük Harf ❌")

    if kontrol_rakam(sifre): # en az 1 rakam şartı varsa onay yoksa error verdirdim
        print("1 Rakam ✅")
    else:
        print("1 Rakam ❌")

    print("\nTekrar Deneyin:") # neyin eksik olduğunu daha açık şekilde gösterdim
    for kural in sonuc:
        print("-", kural)
```
