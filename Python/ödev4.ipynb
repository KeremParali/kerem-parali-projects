```python
#---------- RPG Karakter Filtreleme ---------------------------
# KEREM PARALI 300125024


# Bu ödevde karakter listesi oluşturulup,
# lambda ile bazı kontroller yapıldı,
# for döngüsü ile veriler gezildi,
# ve list comprehension ile filtreleme/etiketleme yapıldı.
# (oyun karakterleri ve özellikleri gerçek bir oyundan esinlenildi)

#--------------------------------------------------------------

karakterler = [ # karakter listesini oluşturdum
    {"isim":"Ahri","sinif":"buyucu","seviye":16,"hp":1400,"altin":480},   # özelliklerini belirledim
    {"isim":"Garen","sinif":"savascı","seviye":18,"hp":2200,"altin":600}, 
    {"isim":"Ashe","sinif":"okcu","seviye":14,"hp":1500,"altin":520},     
    {"isim":"Darius","sinif":"savascı","seviye":17,"hp":2100,"altin":700},
    {"isim":"Jinx","sinif":"okcu","seviye":13,"hp":1300,"altin":450},     
    {"isim":"Lux","sinif":"buyucu","seviye":15,"hp":1350,"altin":550}     
]

# ----------- LAMBDA KISMI ----------- 
# lambda = kontrol fonksiyonu

# okçu mu diye kontrol eden kısa fonksiyon
okcu_mu = lambda x: x["sinif"] == "okcu"

# güçlü mü kontrolü (2 şart birlikte)
guclu_mu = lambda x: (x["seviye"] > 10) and (x["hp"] > 1500) 
# seviyesini ve hp sini lambda ile şartı sağlıyormu kontrol ettik


# okçu şartı kısmı
print("Okçular:") # okçular yazdırdık
for karakter in karakterler: # for in ile listeyi tek tek dolaşıyorum
    if okcu_mu(karakter):   # eğer okçuysa
        print("-", karakter["isim"])  # ismini yaz


print("\nGüçlü olanlar:") # güçlü olma şartı
for karakter in karakterler: # for in ile listeyi dolaşıyoruz
    if guclu_mu(karakter):  # güçlümü kontrol ettik
        print("-", karakter["isim"]) # yazdırdık


# --------- LIST COMPREHENSION ----------

# seviye 15 üstü olanların isimlerini aldırdım
seviye_ustu = [k["isim"] for k in karakterler if k["seviye"] > 15]
# k = değişken adıdır(ahri,garen vs.)

print("\n15 üstü karakterler:", seviye_ustu) # şartı sağlayanları üstü olarak yazdırdım


# altına göre zengin/fakir etiketi veriyorum
etiketler = [] # list 
for k in karakterler: # for in = listeyi tek tek dolaş
    if k["altin"] > 500: # altın 500den büyükse (eğer)
        etiketler.append((k["isim"], "zengin"))  # tuple ekle
    else:
        etiketler.append((k["isim"], "fakir")) # büyükse zengin değilse fakir 

print("\nEtiketler:", etiketler) # tüm bilgileri veren son çıktı
```
